# List Comprehension
Example:

Based on a list of fruits, you want a new list, containing only the fruits with the letter "p" in the name.

```python
l=["Apple","banana","Grapes","melon","orange"]
result =[fruit for fruit in l if "p" in fruit]
result
```

## Example : Only accept items that are not "Apple":

```python
result=[fruit for fruit in l if fruit!="Apple"]
result
```
