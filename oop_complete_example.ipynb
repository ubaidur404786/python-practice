# In this file we will get data from json api 

```python
import requests
class User:
    def __init__(self,id,name,email):
        self.id=id
        self.name=name
        self.email=email
        
    def getUser(self):
        return "id = "+str(self.id) + ", Name = "+self.name +", email = "+ self.email
    
class ControllerAPI:

    def __init__(self):
        self.url=""
    def fetch_data(self):
        data= requests.get(self.url)
        data=data.json()
        return data
    def getUserApi(self,url):
        self.url=url
        data= self.fetch_data()
        users_list=[]
        for i in data:
            users_list.append(User(i["id"],i["name"],i["email"]))
        
        return users_list



```

```python
obj=ControllerAPI()
list=obj.getUserApi("https://jsonplaceholder.typicode.com/users")

for i in list:
    print(i.getUser())


```
