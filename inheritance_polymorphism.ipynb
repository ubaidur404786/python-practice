# Single inheritance with overriding concept

```python
class User:
    def login(self):
        print("User Login page")
    def register(self):
        print("Registration page")

class Student(User):
    def enroll(self):
        print("Enrolled")
    #overriding concept same name of methods in child and parenst 
    def login(self):
        print("student login")
```

```python
stu=Student()
stu.login()
```

# inheritance

```python
class User:
    def login(self):
        print("User Login page")
    def register(self):
        print("Registration page")

class Student(User):
    def enroll(self):
        print("Enrolled")
    #overriding concept same name of methods in child and parenst 
    def login(self):
        print("student login")
class Teacher(User):
    def intruct(self):
        print("Intructed ")
    #overriding concept same name of methods in child and parenst 
    def login(self):
        print("teacher login")
```

```python
stu=Student()
stu.login()

tea=Teacher()
tea.login()


user=User()
user.login()
```

# Method overloading 

in python we use smart way to create overloading methods but it is not same as in java

```python
class Geometry:
    #this is the technique assign 0 to variables
    def area(self,a,b=0):
        if(b==0):
            print("Cricle",3.14 *a*a)
        else:
            print("Rectangular: ",a*b)

obj=Geometry()
obj.area(4)
obj.area(4,5)
# so this is overloading bcz with same methods we are giving different inputs to get a results
```
