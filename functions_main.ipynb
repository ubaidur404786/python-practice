# Basic function examples 


# Tell number is odd or even


```python
def is_even(n):
    if n%2==0:
        return True
    else :
        return False

is_even(5)

```

```python
is_even(4)
```

```python
for i in range(1,11):
    print(str(i) +" is even ? " + str(is_even(i)))
```
