# Factorial problem

```python
def fic(x):
    if(x==1):
        return x
    return x*fic(x-1)

fic(10)
```

# Palindrome problem
## chekc input condition is palindrome or not 
 


```python
def is_palindrome(x):

    if (len(x)<=1):
        return True
    if(x[0]==x[-1]):
        # del x[0],x[-1]
        return is_palindrome(x[1:-1])
    else:
        return False


x=list("madam")
print(is_palindrome(x))
```

```python


print(is_palindrome("abba"))
print(is_palindrome("python"))
```

# Fibonacci Problem 

```python
def fibonacci(x):
    if(x==0 or x==1):
        return 1
    return  fibonacci(x-1)+fibonacci(x-2)

print(fibonacci(12))
```
