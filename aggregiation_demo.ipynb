# Aggregation 
## Example 
### Like Customer has a Address 


```python
class Customer:
    def __init__(self,name,gender,address):
        self.name=name
        self.gender=gender
        self.address=address
class Address:
    def __init__(self,city,pincode,state):
        self.city=city
        self.pincode=pincode
        self.state=state

```

```python
ad1=Address("Nice","123","xyz")
cus1=Customer("Ubaid","M",ad1)


print(cus1.address.city)
```
