```python
class ATM:
    # constructor 
    def __init__(self):
        #encapsulate  data variables
        self.__pin=""
        self.__balance=0

        self.menuATM()

    def menuATM(self):
       


            print("1. Create Pin \n2. Deposit Amount \n3.Withdraw Amount \n4.Check Amount\n5.Exit? ")
            try: 
                choice=int(input("Option ? : "))
            except ValueError:
                print("Invalid data type" )
                
            if(choice==1):
                self.createPin()
            elif(choice==2):
                self.depositAmount()
            elif(choice==3):
                self.withdrawAmount()
            elif(choice==4):
                self.checkAmount()
            elif(choice==5):
                print("Exit!")
                
            else: 
                print("Invalid try again btw 1-5")
            
            


    def createPin(self):
        self.__pin=input("Enter your Pin ")
        print("Pin Create Successfully")

    def depositAmount(self):
        pin=input("Enter your pin : ")
        if(pin==self.__pin):
            self.__balance += int(input("Enter your deposit amount : "))
            print("Deposit Amount Successfully")

    def withdrawAmount(self):
        pin=input("Enter your pin : ")
        if(pin==self.__pin):
            am= int(input("Enter your withdraw amount : "))
            if(am>self.__balance):
                print("Insufficient blance")
            else:
                self.__balance -= am
                print("Deposit Amount Successfully")

    def checkAmount(self):
        pin=input("Enter your pin : ")
        if(pin==self.__pin):
            print("Balance is : " , self.__balance)
    

    
```
